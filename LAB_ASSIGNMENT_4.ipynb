# **Q2) Read iris.csv using pandas dataframe, next perform Histogram, Line plot, and box plot over the dataset.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
data = pd.read_csv("iris.csv")
```

```python
plt.figure(figsize = (10, 7))
x = data["sepal.length"]
  
plt.hist(x, bins = 20, color = "green")
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
```

```python
plt.figure(figsize = (10, 7))
x = data["petal.length"]
  
plt.hist(x, bins = 20, color = "green")
plt.title("Petal Length in cm")
plt.xlabel("Petal_Length_cm")
plt.ylabel("Count")
  
plt.show()
```

```python
plt.plot(data["sepal.length"], "r--")
plt.show()
```

```python
new_data = data[["sepal.length", "sepal.width", "petal.length", "petal.width"]]
print(new_data.head())
```

```python
plt.figure(figsize = (10, 7))
new_data.boxplot()
```

# Q3) Lets assume pre = np.array([19, 6, 11, 9]) and post = np.array([15, 11, 9, 8]) are the two Numpy array, plot this two array as side by side bar plot.

```python
pre = np.array([19, 6, 11, 9])
post = np.array([15, 11, 9, 8])
labels=['Survey '+x for x in list('ABCD')]
# the plot – left then right
fig, ax = plt.subplots(figsize=(8, 3.5))
width = 0.4 # bar width
xlocs = np.arange(len(pre))
ax.bar(xlocs-width, pre, width, color='green', label='True')
ax.bar(xlocs, post, width, color='#1f10ed', label='False')
# labels, and title, then save
ax.set_xticks(ticks=range(len(pre)))
ax.set_xticklabels(labels)
ax.legend(loc='best')
ax.set_ylabel('Count')
fig.suptitle('Sample Chart')
fig.savefig('filename.png', dpi=125)
```

# Q4) Generate three NumPy arrays (X1, X2, X3) by getting 20 equidistance points between 0-100.Next convert this array into Y1, Y2, Y3 by applying sin, cos, and tan function over X1, X2, X3, respectively. Finally, plot these multiple lines over the same axis on a line plot graph.

```python
import numpy as np
import pandas as pd
# generate sample data for multiple line
x1 = np.linspace(0, 100, 20)
x2 = np.linspace(0, 100, 20)
x3 = np.linspace(0, 100, 20)
y1 = np.sin(x1)
y2 = np.cos(x2)
y3 = np.tan(x3)
# get the Figure and Axes all at once using subplots
fig, ax = plt.subplots(figsize=(8,4))
# line plot data on the Axes
ax.plot(x1, y1, label='sin')
ax.plot(x2, y2, label='cos')
ax.plot(x3, y3, label='tan')
# add legend
ax.legend(loc='best', prop={'size':'large'})
```

# Q5) Write a python program: Create list x = [5, 2, 9, 4, 7] and y = [10, 5, 8, 4, 2] : perform line,scatter functions over the list x and y. b) Plotting two or more lines on the same plot Y1=[8,6,7,5,1]. c) Naming the x and y-axis. Giving a title to my graph. Show a legend on the plot.

```python
X=[5,2,9,4,7]
Y=[10,5,8,4,2]
plt.plot(X,Y)
plt.show()

plt.scatter(X,Y)
plt.show()
```

```python
Y1=[8,6,7,5,1]
# plotting the line 2 points 
plt.plot(X, Y, label = "line 1")
plt.plot(X, Y1, label = "line 2")
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title of the current axes.
plt.title('Two or more lines on same plot with suitable legends ')


# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```
