```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 3) Read a CSV datafile using pandas dataframe then apply 1) min-max normalization and 2)z-score normalization over the each column of the dataset.

```python
from sklearn import datasets
from sklearn import preprocessing
# Load dataset from sklearn dataset package
iris = datasets.load_iris()
# seperate input and target features
X = iris.data
y = iris.target
print(X)
#Min-max normalization
minmax_scale = preprocessing.MinMaxScaler().fit(X)
X_minmax = minmax_scale.transform(X)
print(X_minmax)
```

```python
iris = datasets.load_iris()
# seperate input and target features
X = iris.data
y = iris.target
print(X)
#Z-score normalization
std_scale = preprocessing.StandardScaler().fit(X)
X_std = std_scale.transform(X)
print(X_std)
```
